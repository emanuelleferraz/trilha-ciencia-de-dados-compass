```python
import pandas as pd
from apyori import apriori
```

## Base de dados mercado 1

```python
base_mercado1 = pd.read_csv('mercado.csv', header = None)
base_mercado1
```

```python
base_mercado1.shape
```

### Transformação dos Dados

 O algoritmo Apriori requer que os dados estejam em formato de lista.


```python
transacoes = []
for i in range(len(base_mercado1)):
  transacoes.append([str(base_mercado1.values[i, j]) for j in range(base_mercado1.shape[1])]) #shape[1] = colunas
```

```python
transacoes
```

```python
type(transacoes)
```

### Regras e Funcionamento do Algoritmo

```python
regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.8, min_lift = 2)
resultados = list(regras)
```

```python
A = [] # se
B = [] # então
suporte = []
confianca = []
lift = []

for resultado in resultados:
  s = resultado[1]
  result_rules = resultado[2]
  for result_rule in result_rules:
    a = list(result_rule[0])
    b = list(result_rule[1])
    c = result_rule[2]
    l = result_rule[3]
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
```

## Base de dados mercado 2

```python
base_mercado2.shape
```

```python
transacoes = []
for i in range(base_mercado2.shape[0]):
  transacoes.append([str(base_mercado2.values[i, j]) for j in range(base_mercado2.shape[1])])
```

```python
transacoes
```

```python
# Para definir o suporte, é interssante definir também um objetivo, por exemplo:
# Produtos que são vendidos 4 vezes por dia -> 4 * 7 = 28
```

```python
# Suporte
28 / 7501
```

### Regras de Associação e Crição do novo Dataframe

```python
regras = apriori(transacoes, min_support = 0.003, min_confidence = 0.2, min_lift = 3)
resultados = list(regras)
len(resultados)
```

Dificilmente você conseguirá gerar regras com suporte de 30% em base de dados muito grandes. O mesmo serve para a confiança.

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
  s = resultado[1]
  result_rules = resultado[2]
  for result_rule in result_rules:
    a = list(result_rule[0])
    b = list(result_rule[1])
    c = result_rule[2]
    l = result_rule[3]
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
rules_df.sort_values(by = 'confianca', ascending = False)
```
