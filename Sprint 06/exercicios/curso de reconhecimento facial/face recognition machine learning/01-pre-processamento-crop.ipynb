# Pré Processamento dos Dados - Crop Faces

# Importações

```python
import numpy as np 
import cv2 
from glob import glob 
import matplotlib.pyplot as plt 
```

```python
# Importando imagens
fpath = glob('./data/female/*.jpg')
mpath = glob('./data/male/*.jpg')
```

```python
print('O número de imagens na pasta Feminino = ',len(fpath))
print('O número de imagens na pasta Masculino = ',len(mpath))
```

# Identificando e Cortando os Rostos com CascadeClassifier

```python
# Lendo as imagens e covertendo para RGB
img = cv2.imread(fpath[2]) 
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# Aplicando o CascadeClassifier
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY) # gray scale
faces_list = haar.detectMultiScale(gray,1.5,5)
for x,y,w,h in faces_list:
    cv2.rectangle(img_rgb,(x,y),(x+w,y+h),(0,255,0),2) # Cortando os rostos
    roi = img_rgb[y:y+h,x:x+w]
    
    plt.imshow(roi)
    plt.axis('off')
    plt.show()

# Salvando a imagem
```

# Cortando as Imagens

 - Cortando todas as imagens femininas

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')

# Aplicando os mesmos passos acima
for i in range(len(fpath)):
    try:
        img = cv2.imread(fpath[i]) 
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 

        gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray,1.5,5)
        for x,y,w,h in faces_list:
            roi = img[y:y+h,x:x+w]
            cv2.imwrite(f'./crop_data/female/female_{i}.jpg',roi)
            print('Imagem processada com sucesso!')
    except:
        print('Não foi possível processar a imagem.')
```

 - Cortando todas as imagens masculinhas

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')

# Aplicando os mesmos passos acima
for i in range(len(mpath)):
    try:
        img = cv2.imread(mpath[i]) 
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 

        gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray,1.5,5)
        for x,y,w,h in faces_list:
            roi = img[y:y+h,x:x+w]
            cv2.imwrite(f'./crop_data/male/male_{i}.jpg',roi)
            print('Imagem processada com sucesso!')
    except:
        print('Não foi possível processar a imagem.')
```
