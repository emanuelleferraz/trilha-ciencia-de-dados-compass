# Importações

```python
import numpy as np
import cv2
import matplotlib.image as matimg
from PIL import Image

import matplotlib.pyplot as plt
```

# Usando OpenCV, Matplot e Pillow

```python
# OpenCV - BGR (Blue, Green, Red)
img_cv = cv2.imread('./data/flying_bird.jpg')

# Matplotlib - RGB (Red, Green, Blue)
img_mat = matimg.imread('./data/flying_bird.jpg')

# Pillow - RGB
img_pil = Image.open('./data/flying_bird.jpg')
```

```python
# OpenCV - array com valores dos pixels (BGR)
print(img_cv)
```

```python
# Matplotlib - array com valores dos pixels (RGB)
print(img_mat)
```

```python
# Pillow
img_pil
```

```python
# Transformando em array
img_pill_array = np.array(img_pil)
```

```python
img_pill_array
```

# Display

```python
# Display com OpenCV
cv2.imshow('OpenCV BGR', img_cv)
cv2.imshow('Matplotlib RGB', img_mat)
cv2.imshow('Pillow RGB', img_pill_array)
cv2.waitKey(0) # para fechar a janela
cv2.destroyAllWindows()
```

```python
# Display com Matplotlib
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')

plt.subplot(1,3,2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')

plt.subplot(1,3,3)
plt.imshow(img_pill_array)
plt.axis('off')
plt.title('Pillow RGB')

plt.show()
```

# Color Space

```python
b = img_cv[:,:,0]
g = img_cv[:,:,1]
r = img_cv[:,:,2]
```

```python
cv2.imshow('blue channel',b)
cv2.imshow('green channel',g)
cv2.imshow('red channel',r)

cv2.imshow('BGR',img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Usando o split
b,g,r  = cv2.split(img_cv)
```

```python
cv2.imshow('blue channel',b)
cv2.imshow('green channel',g)
cv2.imshow('red channel',r)

cv2.imshow('BGR',img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Convertendo BGR em RGB

```python
img_cv_rgb = cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cv_rgb)
plt.axis('off')
plt.show()
```

# Gray Scale

```python
img_gray = cv2.cvtColor(img_cv,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('grayscale',img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
