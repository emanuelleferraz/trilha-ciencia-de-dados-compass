## Amostragem Estratificada

```python
enem_ribeirao_preto['RACA'].value_counts()
```

 - Para uma quantidade pequena de grupos

```python
branca = enem_ribeirao_preto.loc[enem_ribeirao_preto.RACA == 'branca'].sample(frac=0.10)
branca.shape
```

```python
parda = enem_ribeirao_preto.loc[enem_ribeirao_preto.RACA == 'parda'].sample(frac=0.10)
parda.shape
```

```python
preta = enem_ribeirao_preto.loc[enem_ribeirao_preto.RACA == 'preta'].sample(frac=0.10)
preta.shape
```

```python
amarela = enem_ribeirao_preto.loc[enem_ribeirao_preto.RACA == 'amarela'].sample(frac=0.10)
amarela.shape
```

```python
nao_declarado = enem_ribeirao_preto.loc[enem_ribeirao_preto.RACA == 'nao_declarado'].sample(frac=0.10)
nao_declarado.shape
```

```python
indigena = enem_ribeirao_preto.loc[enem_ribeirao_preto.RACA == 'indigena'].sample(frac=0.10)
indigena.shape
```

```python
# Juntanto todos esses dados em uma única tabela
amostra_estratificada = pd.concat([branca, parda, preta, amarela, nao_declarado, indigena])
```

```python
amostra_estratificada
```
