- Criando função de probabilidade

```python
# Função de Probabilidade

def probabilidade(A, E):
    resultado = (A / E)*100
    print(f"{resultado:.2f}")
```

```python
# Probabilidade de retirar um vestibulando de Osasco:
probabilidade(len(enem_osasco), len(enem_sp))
```

```python
# Probabilidade de retirar um vestibulando de Santo André:
probabilidade(len(enem_santo_andre), len(enem_sp))
```

- Criando função de probabilidade de NÃO ocorrer um evento

```python
def probabilidade_nao_ocorrer_evento(A, E):
    resultado = (1 - (A/E))*100
    print(f"{resultado:.2f}")
```

```python
# Probabilidade de não retirar um vestibulando de Osasco
probabilidade_nao_ocorrer_evento(len(enem_osasco), len(enem_sp))
```

```python
# Probabilidade de não retirar um vestibulando de Santo André
probabilidade_nao_ocorrer_evento(len(enem_santo_andre), len(enem_sp))
```

```python
# Probabilidade de não retirar um vestibulando NEM de Osasco NEM de Santo André
probabilidade_nao_ocorrer_evento(len(enem_osasco)+len(enem_santo_andre), len(enem_sp))
```

## Probabilidade da União Mutuamente Exclusivos (A U B)

- Criando função probabilidade da união mutuamente exclusivos

```python
# Função
def probabilidade_uniao_exclusivo(A, B, E):
    resultado = (A/E + B/E)*100
    print(f"{resultado:.2f}")
```

```python
# Probabilidade de retirar um vestibulando de Osasco OU de Santo André
probabilidade_uniao_exclusivo(len(enem_osasco), len(enem_santo_andre), len(enem_sp))
```

## Probabilidade da Intersecção de dois Eventos

```python
# Função
def probabilidade_interseccao(A, B, E):
    resultado = (A/E * B/E)*100
    print(f"{resultado:.2f}")
```

```python
# Probabilidade de retirar um vestibulando de Osasco E de Santo André (com reposição)
probabilidade_interseccao(len(enem_osasco), len(enem_santo_andre), len(enem_sp))
```

## Probabilidade Condicional

```python
# Função
def probabilidade_condicional(AB, B):
    resultado = (AB/B)*100
    print(f"{resultado:.2f}")
```

```python
# Probabilidade de retirar uma mulher parda
mulher = enem_sp.loc[enem_sp.SEXO == 'F']
```

```python
# Mulher dada a condição de que ela seja parda
mulher_parda = mulher.loc[mulher.RACA == 'parda']
```

```python
mulher_parda.head(10)
```

```python
# Probabilidade de retirar uma vestibulanda PARDA
probabilidade_condicional(len(mulher_parda), len(mulher))
```

```python
mulher_escola_publica = mulher.loc[mulher.ESCOLA == 'pública']
```
