```python
import numpy as np
import pandas as pd
```

```python
dados = pd.read_csv('microdados_enem_2019_sp.csv', sep=';', encoding='iso-8859-1')
```

### Extração e Limpeza dos Dados

```python
# Visualização
dados.head()
```

```python
# linhas e colunas (Linhas serão a quantidade de inscritos)
dados.shape
```

```python
# análise dos tipos de atributos
dados.dtypes
```

```python
# Excluindo uma coluna (colunas que não são interessantes para análise)
dados1 = dados.drop(columns=['CO_MUNICIPIO_RESIDENCIA'])
```

```python
# Conferindo se a exclusão deu certo
dados1.head()
```

```python
# Exclusão de outras colunas (presenã do , no final ao cortar mais de uma coluna)
dados1 = dados1.drop(columns=['CO_UF_RESIDENCIA','SG_UF_RESIDENCIA','CO_MUNICIPIO_NASCIMENTO'], )
```

```python
dados1.head(15)
```

```python
# Excluindo outras colunas que não serão utilizadas no estudo
dados1 = dados1.drop(columns=['NO_MUNICIPIO_NASCIMENTO','CO_UF_NASCIMENTO','SG_UF_NASCIMENTO',
                              'TP_ANO_CONCLUIU','TP_ENSINO','CO_MUNICIPIO_ESC','CO_UF_ESC','SG_UF_ESC'],)
```

```python
dados1.head()
```

```python
dados1 = dados1.drop(columns=['TP_DEPENDENCIA_ADM_ESC','TP_LOCALIZACAO_ESC','TP_SIT_FUNC_ESC'],)
```

```python
dados1.head()
```

```python
dados1.shape
```

```python
# Correção do erro nas notas das disciplinas
# Como corrigir? 
# Dividir todas as colunas dessas disciplinas por 10

dados1.loc[:, 'NU_NOTA_CN'] /=10
```

```python
dados1.head()
```

```python
# Corrigindo de todas
dados1.loc[:, 'NU_NOTA_CH'] /=10
dados1.loc[:, 'NU_NOTA_LC'] /=10
dados1.loc[:, 'NU_NOTA_MT'] /=10
```

```python
dados1.head()
```

```python
# Renomeando os nomes das colunas
dados1 = dados1.rename(columns={'NU_NOTA_REDACAO': 'NOTA_REDACAO'})
```

```python
dados1.head()
```

```python
# Renomeando nome das colunas de notas das disciplinas
dados1 = dados1.rename(columns={'NU_NOTA_CN': 'NOTA_CN','NU_NOTA_CH': 'NOTA_CH',
                                'NU_NOTA_LC': 'NOTA_LC','NU_NOTA_MT': 'NOTA_MT'})
```

```python
# Renomeando nomes das colunas de notas das competencias
dados1 = dados1.rename(columns={'NU_NOTA_COMP1': 'COMP1','NU_NOTA_COMP2': 'COMP2',
                                'NU_NOTA_COMP3': 'COMP3','NU_NOTA_COMP4': 'COMP4',
                                'NU_NOTA_COMP5':'COMP5'})
```

```python
# Renomeando nome das colunas de dados pessoais
dados1 = dados1.rename(columns={'NU_IDADE': 'IDADE','TP_SEXO': 'SEXO',
                                'TP_COR_RACA': 'RACA','Q025': 'INTERNET',
                                'TP_ESCOLA':'ESCOLA'})
```

```python
dados1.head()
```

```python
# Renomeando linhas de um dataframe
# Renomeando raça
dados1["RACA"] = dados1["RACA"].replace({0:"nao_declarado", 1:"branca", 2:"preta", 3:"parda", 4:"amarela", 5:"indigena"})
```

```python
# Renomeando lingua estrangeira
dados1["TP_LINGUA"] = dados1["TP_LINGUA"].replace({0:"Inglês", 1:"Espanhol"})
```

```python
# Renomeando tipo de escola
dados1["ESCOLA"] = dados1["ESCOLA"].replace({1:"não_respondeu", 2:"pública", 3:"privada", 4:"exterior"})
```

```python
# Renomeando questão Q25 sobre internet
dados1["INTERNET"] = dados1["INTERNET"].replace({"A":"Não", "B":"Sim"})
```

```python
dados1.head()
```
