# Distribuição de Frequências no Python

```python
import numpy as np 
import pandas as pd
from collections import Counter
```

```python
enem_sp = pd.read_csv('enem_2019_tratado.csv', sep=',', encoding='iso-8859-1')
```

```python
enem_sp.head()
```

Trabalhando com outra cidade, o professor usa Sorocaba, mas irei usar Guarulhos.

```python
enem_guarulhos = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Guarulhos']
```

```python
enem_guarulhos.head()
```

```python
enem_guarulhos.shape
```

```python
idades = enem_guarulhos['IDADE']
```

```python
idades
```

```python
frequencia_absoluta = Counter(idades)
```

```python
frequencia_absoluta
```

```python
frequencia_absoluta = pd.DataFrame.from_dict(frequencia_absoluta, orient='index')
```

```python
frequencia_absoluta
```

```python
frequencia_absoluta = frequencia_absoluta.sort_index(ascending=True)
frequencia_absoluta
```

```python
frequencia_relativa = frequencia_absoluta / frequencia_absoluta.sum()
frequencia_relativa
```

```python
# Em porcentagem
frequencia_relativa_perc = frequencia_relativa * 100
frequencia_relativa_perc
```

# Histogramas


```python
import matplotlib.pyplot as plt
```

```python
enem_guarulhos.IDADE.hist(bins=9) #bins está relacionado a faixa de valores
plt.title('Distribuição de Idades - Enem SP 2019 - Guarulhos')
plt.show()
```

```python
# Utilizando matplotlib
plt.hist(enem_guarulhos.IDADE, 8, rwidth=1)
plt.title('Distribuição de Idades - Enem SP 2019 - Guarulhos')
plt.show()
```

```python
# Melhorando o gráfico
from matplotlib import colors
plt.hist(enem_guarulhos.IDADE, 10, rwidth=1, color='gray', density = True, )
plt.title('Distribuição de Idades - Enem SP 2019 - Guarulhos')
plt.xlabel ('Idade')
plt.ylabel ('Frequência')
plt.show()
```
