# Exercícios sobre Gráficos - Pandas

```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
# Abrindo arquivo
data_arq = 'data/games.csv'
dataset = pd.read_csv(data_arq)
```

```python
dataset = dataset.dropna()
```

```python
dataset
```

```python
# Gerando Subplots
dataset.plot(subplots=True, layout=(3, 3), figsize=(10,10), kind="hist")
```

```python
dataset["Platform"].unique()
```

```python
# Quantidade de Jogos por plataforma
dataset.groupby("Platform")["Platform"].count().plot(kind="barh")
```

```python
# Usando gráfico de Pizza
dataset.groupby("Platform")["Platform"].count().plot(kind="pie")
```

```python
dataset.columns
```

```python
# Criando um gráfico de dispersão entre Critic_Score e Global Sales
dataset.plot.scatter(x = "Critic_Score", y = "Global_Sales")
```

```python
# Criando histograma da coluna Critic_Score
dataset["Critic_Score"].plot.hist()
```
