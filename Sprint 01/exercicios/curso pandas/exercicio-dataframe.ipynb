```python
# Tipos de dados
dataset.dtypes
```

```python
# Convertendo os tipos
dataset = dataset.astype({"Year_of_Release": "int", "Critic_Count": "int", "User_Score": "float", "User_Count": "int" })
```

```python
dataset.dtypes
```

```python
# Função para atualizar coluna
def atualiza_user_score(score):
    return score * 10
```

```python
# Atualizando coluna usando apply
dataset["User_Score"] = dataset["User_Score"].apply(atualiza_user_score)
```

```python
dataset
```

```python
# Visualização de linhas e colunas
dataset.iloc[0:5, 0:5]
```

```python
# Apagando colunas
dataset.drop(["NA_Sales", "JP_Sales", "EU_Sales", "Critic_Count", "User_Count"], axis=1, inplace=True)
```

```python
dataset
```

```python
# Retornando valores das colunas
for coluna in dataset.columns:
    if dataset[coluna].dtype == object:
        print("----------------------------", coluna, "----------------------------")
        print(dataset[coluna].value_counts())
        print
```

```python
# Retornando porcentagem
for coluna in dataset.columns:
    if dataset[coluna].dtype == object:
        print("----------------------------", coluna, "----------------------------")
        print(dataset[coluna].value_counts(normalize=True))
        print()
```

```python
# Retornando os 5 jogos mais vendidos
dataset.sort_values(["Global_Sales"], ascending=False)
```

```python
# Retornando os 5 jogos menos vencidos
dataset.sort_values(["Global_Sales"], ascending=True)
```

```python
# Retornando todos os jogos com colunas específicas
dataset.loc[dataset["Platform"] == "PS4", ["Name", "Year_of_Release", "Developer"]]
```

```python
# Jogos de PS4 com avaliação de usuário maior que 90
dataset.loc[(dataset["Platform"] == "PS4") & (dataset["User_Score"] > 90) ]
```

```python
# Agora usando query()
dataset.query("Platform == 'PS4' and `User_Score` > 85")
```

```python
# Inserindo nova coluna (média entre scores de critica e usuários)
dataset["Global_Score"] = (dataset["Critic_Score"] + dataset["User_Score"]) / 2 
```

```python
dataset.head()
```

```python
# Apagando colunas de score
dataset.drop(["Critic_Score", "User_Score"], axis=1, inplace=True)
```

```python
dataset.head()
```

```python
# Reordenando as colunas
dataset.columns
```

```python
dataset = dataset.reindex(labels=['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher',
       'Other_Sales', 'Global_Sales', 'Global_Score', 'Developer', 'Rating'], axis=1)
```

```python
dataset
```

```python
# Alterando as colunas categóricas para category
dataset = dataset.astype({"Name": "category", "Platform": "category", "Genre": "category", "Publisher": "category", "Developer": "category", "Rating": "category"})
```

```python
dataset.info()
```

```python
# Somatório dos registros de vendas globais
dataset.loc[:, ["Global_Sales"]].sum()
```

```python
# Média dos registros de vendas globais
dataset.loc[:, ["Global_Sales"]].mean()
```

```python
# Valores unicos de Platform
dataset["Platform"].unique()
```

```python
# Retornando agrupamento de Platform por vendas globais
dataset.groupby("Platform")["Global_Sales"].mean().sort_values(ascending=False)
```

```python
# Agrupamento de Genre e Platform + média dos atributos
dataset.groupby(["Platform", "Genre"]).agg("mean", numeric_only=True)
```

```python
# Contagem de vendas definidas por gênero
dataset.groupby("Genre")["Global_Sales"].count()
```

```python
# A média de Global Sales definida pelo Genre
group = dataset.groupby(["Genre"], as_index=False)["Global_Sales"].mean()
group
```

```python
dataset.columns
```

```python
# Nova coluna com média do Score Global
group.assign(avg_score = dataset.groupby(["Genre"])["Global_Score"].transform("mean"))
```

```python
# Criando uma tabela pivÔ
dataset.pivot_table(index="Genre", columns="Platform", values="Global_Sales", aggfunc="mean")
```
