## Criação de Dataframes

```python
# Criando dataframe
data = [['Alice', 25, 'São Paulo'],
        ['João', 30, 'Rio de Janeiro'],
        ['Bruno', 35, 'Curitiba'],
        ['Carlos', 40, 'Manaus']]
```

```python
type(data)
```

```python
# Covertendo a lista para dataframe
# Criando colunas
colunas = ['Nome', 'Idade', 'Cidade']
```

```python
# Criando dataframe
df = pd.DataFrame(data, columns=colunas)
df
```

```python
# Criando dataframes utilizando dicionários
data_dic = {'Nome': ['Alice', 'João', 'Bruno', 'Carlos'],
            'Idade': [25, 30, 35, 40],
            'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Manaus']}
data_dic
```

```python
# Criação do dataframe
dfc = pd.DataFrame(data_dic)
dfc
```

```python
# Outra forma de criar dataframe
# np array
data = np.array([['Alice', 25, 'São Paulo'],
        ['João', 30, 'Rio de Janeiro'],
        ['Bruno', 35, 'Curitiba'],
        ['Carlos', 40, 'Manaus']])
#type(data)
#data

# Criando dataframe
df = pd.DataFrame(data, columns=colunas)
df
```

```python
# Criando dataframe usando tuplas
data_tup = [('Alice', 25, 'São Paulo'),
        ('João', 30, 'Rio de Janeiro'),
        ('Bruno', 35, 'Curitiba'),
        ('Carlos', 40, 'Manaus')]

#type(data_tup), type(data_tup[0])
#data_tup

# Criando dataframe
dft = pd.DataFrame(data_tup, columns=colunas)
dft
```

Um dataframe é um conjunto de várias Series. Cada uma das colunas é considerada uma Serie.

```python
# Definindo uma Serie
data = {'Nome': pd.Series(['Alice', 'João', 'Bruno', 'Carlos']),
        'Idade': pd.Series([25, 30, 35, 40]),
        'Cidade': pd.Series(['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Manaus'])}
#data

# Criação do dataframe
dfs = pd.DataFrame(data)
dfs
```

```python
# Dicionário com numpy array
data = {'Nome': np.array(['Alice', 'João', 'Bruno', 'Carlos']),
        'Idade': np.array([25, 30, 35, 40]),
        'Cidade': np.array(['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Manaus'])}
#data
#type(data)

# Criação do dataframe
dfn = pd.DataFrame(data)
dfn
```

## Valores Faltantes

```python
dataset.isna().sum() # não é NaN -> '?'
```

```python
# Teste
dataset.loc[dataset["workclass"].str.contains("\?")]
```

```python
dataset["workclass"].str.contains("\?").sum()
```

```python
colunas_categoricas = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']

for coluna in colunas_categoricas:
  #print(coluna)
  tem_interrogacao = dataset[coluna].str.contains("\?").any()
  if tem_interrogacao:
    print(f"Na coluna '{coluna}' existem valores '?'")
```

```python
# Substituindo ? pelo NaN
for coluna in colunas_categoricas:
    dataset[coluna] = dataset[coluna].replace(' ?', np.nan)
```

```python
# Quantidade de valores faltantes
dataset.isna().sum()
```

```python
# Maneira mais rápida de fazer a substituição dos valores faltantes
dataset.fillna("Not Informed").iloc[0:30]
```

```python
# Uma outra opção é fazer a extração da moda
dataset["workclass"].mode() 
```

```python
dataset["occupation"].mode() 
```

```python
dataset["native-country"].mode() 
```

```python
# Substituindo
dataset["workclass"].fillna(dataset["workclass"].mode().iloc[0], inplace=True)
```

```python
dataset["occupation"].fillna(dataset["occupation"].mode().iloc[0], inplace=True)
```

```python
dataset["native-country"].fillna(dataset["native-country"].mode().iloc[0], inplace=True)
```

```python
# Agora todos os valores faltantes estão preenchidos. Tratamento feito!
dataset.isna().sum()
```

```python
# Você pode apagar os registros com valores faltantes
dataset.dropna(subset="workclass")
```

## Concatenação

```python
data = {'idvenda': [123, 374, 654, 345],
        'data': ['2024-01-19', '2024-01-20', '2024-02-21', '2024-02-25'],
        'idvendedor': [1, np.nan, 2, 3],
        'valor_total': [45.76, 102.34, 56.34, 34.21]}
df1 = pd.DataFrame.from_dict(data)
df1
```

```python
data = {'idvenda': [546, 232, 789, 2345],
        'data': ['2024-01-22', '2024-01-22', '2024-01-23', '2024-01-24'],
        'idvendedor': [3, 2, 1, 2],
        'valor_total': [45.76, 102.34, 56.34, 34.21]}
df2 = pd.DataFrame.from_dict(data)
df2
```

```python
# Concatenando
data_complete = pd.concat([df1, df2], ignore_index = True)
data_complete
```

```python
data_complete.query("idvendedor == 1")
```

## Junção

```python
# Dataframe do vendedor
data = {'idvendedor': [1, 2, 3, 4],
        'nome': ['Julia', 'Paulo', 'Jéssica', 'Marcos']}
df_vendedor = pd.DataFrame.from_dict(data)
df_vendedor
```

```python
# Fazendo a junção do Dataframe de vendas com de vendedor
data_complete.merge(df_vendedor, how = "inner", left_on = ["idvendedor"], 
                    right_on = ["idvendedor"])
```

```python
data_complete = data_complete.merge(df_vendedor, how = "left", 
                    left_on = ["idvendedor"],
                    right_on = ["idvendedor"])
```

```python
data_complete
```

```python
data_complete.query("nome == 'Paulo'")
```
